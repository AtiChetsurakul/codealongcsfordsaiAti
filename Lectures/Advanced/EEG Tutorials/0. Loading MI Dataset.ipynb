### 0. Loading MI Dataset

References: https://braindecode.org/auto_examples/plot_dataset_example.html#

In this tutorial, we gonna use http://www.bbci.de/competition/iv/ 2a dataset which is a motor imagery data, where participants were asked to imagine left and right movements. 

To easily load it, we will use braindecode library which has already provide a simple API for loading the dataset.

```python
#pip install braindecode, moabb
from braindecode.datasets.moabb import MOABBDataset

subject_id = [1,2,3,4,5,6,7,8,9]
#subject_id = [1]
#2a= BNCI2014001 2b= BNCI2014004
dataset = MOABBDataset(dataset_name="BNCI2014001", subject_ids=subject_id)
print(dataset)
```

What we get is some matlab file.  To easily use with Python MNE library, we shall convert it to .fif file.  This can be easily done by using the <code>.save</code> function.

```python
dataset.save("data")
```
